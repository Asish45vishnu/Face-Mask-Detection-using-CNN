```python

from keras.models import load_model
import cv2
import numpy as np

```

```python
asish_model = load_model("D://1vishnu_vit//ai\Face mask detection with cnn//face_detection.h5")
face_clsfr = cv2.CascadeClassifier("D://1vishnu_vit//ai//Face mask detection with cnn//haarcascade_frontalface_default.xml")
source = cv2.VideoCapture(0)
labels_dict ={0:'Mask present',1:'No Mask'}
color_dict = {0:(0,255,0),1:(0,0,255)}
while(True):
  ret,img = source.read()
  faces = face_clsfr.detectMultiScale(img,scaleFactor= 1.1,minNeighbors=5)

  for x,y,w,h in faces:
    face_img = img[y:y+w,x:x+w]
    face_img1 = cv2.cvtColor(face_img,cv2.COLOR_BGR2RGB)
    resized = cv2.resize(face_img1,(120,120))
    normalized=resized/255.0
    reshaped=np.reshape(normalized,(1,120,120,3))
    result=asish_model.predict(reshaped)


    label=np.argmax(result,axis=1)[0]

    cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)
    cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
    cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)


    cv2.imshow('LIVE_FACE_DETECTION',img)
    key=cv2.waitKey(1)

    if(key==27):
        break

cv2.destroyAllWindows()
source.release()
```
